## ETS Decomposition of Airline Passengers Dataset:

## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
```

## Read dataset

```python
traffic = pd.read_csv('traffic.csv', index_col='DateTime',parse_dates=True)
```

```python
traffic.drop('ID', axis=1, inplace=True)
```

## show dataset

```python
traffic.head()
```

## Apply ETS

```python
result = seasonal_decompose(traffic['Vehicles'], model='additive',period=48)
```

## Plot ETS

```python
plt.figure(figsize=(22,22))
result.plot()
plt.show()
```
