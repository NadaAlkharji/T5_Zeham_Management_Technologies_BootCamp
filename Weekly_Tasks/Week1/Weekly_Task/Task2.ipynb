# Tasks

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Task (2) Dataset.csv')
```

```python
df2 = pd.read_csv('Task (1) Dataset.csv')
```

```python
ctm = df2[['Weather_conditions']]
```

```python
df = pd.concat([df, ctm], axis=1 )
```

### Identify and analyze patterns in the data.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['TravelTimeLive'], kde=True, bins=20, color='blue')
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['JamsCount'], kde=True, bins=20, color='blue')
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['JamsLength'], kde=True, bins=20, color='blue')
```

```python
sns.histplot(df['JamsDelay'], kde=True, color='blue')
```
